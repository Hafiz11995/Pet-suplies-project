### DATA VALIDATION
The original data is **1500 rows** and **8 columns**. I will explain what I did by each columns in cleaning below;

- **product_id:** This is the unique identifier of the product, there were no missing values as it’s impossible due to the database structure. It’s exactly as expected.
- **category:** This is the category of the products, which includes Housing, Food, Toys, Equipment, Medicine and Accessory. But this field had some errors, it contain some records **“-“** which is not part of the category given. There are **25 values** that contained -. This - make is difficult to get the correct data type, hence I had to replace all the value that contained - with the overall “Unknown”. We now have 7 different category. There are no missing values. This field contained 1500 distinct values and 7 unique values. 
- **animal:** This is the type of animal the product is for. The type of animal are Dog, Cat, Fish and Bird. This field contained no errors, and there are no missing values. There are **1500 distinct values** and **4 unique values**.
- **size:** In this field, there were errors as some values have the same meaning but they were separated by various different capitalizations. Further analysis showed that this was due to inconsistency in size name. For instance, we have **9 different sizes** which was analyzed. In order to correct this, I converted all of them to **lower cases** leaving just **3 different** sizes as provided. There are no missing value in this field. 
- **price:** This is the price the product is sold at. The datatype was an object type and I had to change it to a numeric type because the field contain some record **“NA”** which is an error as it’s impossible to sell a product without a price. I replaced the value that contained NA with the **“overall median price”** which was **29.42** and rounded the values in the field to **2 decimal place**. There were no missing values. The number of NA before the replacing was **150**.
- **sales:** This is the value of all sales of the product in the last year. There were no any missing value and contained no errors. I rounded the values in the field to **2 decimal place**.
- **rating:**  This is the customer rating of the product from **1 to 10**. There was **150 missing values** in this field. I replaced those missing values with **zero**. This field contained **1500 distinct values** and **11 unique values**.
- **repeat_purchases:** This means whether customers repeatedly buy the product **(1)** or not **(0)**. There are no missing value here, hence I didn’t remove any row. There are **150 distinct values** and **2 unique values**.
